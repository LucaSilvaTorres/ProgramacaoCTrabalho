```python
medicamento = []
dia = []
hora = []
quantidade = []
acao = input("1 - Cadastrar| 2 - Consultar medicamento ou dia| 3 - Medicamentos cadastrados| 4 - Sair")
while acao != "4" and acao != "":
    if acao == '1':
        item = input("Informe o medicamento")   
        medicamento.append(item)
        dias = input("Informe o dia")
        dia.append(dias)
        horas = input("Informe a hora")
        hora.append(horas)
        quantidades = input("Informe a quantidade")
        quantidade.append(quantidades)
        print("{}: {} às {} - {} comprimidos - CADASTRADO!".format(item, dias, horas, quantidades))
    elif acao == "2":
        acao2 = input("1 - Consultar por remédio|2 - Consultar por dia: ")
        if acao2 == "1":
            print("Medicamentos disponíveis: ")
            texto0 = set(medicamento)
            textoMedicamento = "\n".join(texto0)
            print(textoMedicamento)
            item2 = input("Informe o medicamento para consultar: ")  
            if item2 in textoMedicamento:
                print(f"{item2}:")
                for i, itens in enumerate(medicamento):
                    if itens == item2:
                        print("\t{} às {} - {} comprimidos".format(dia[i], hora[i], quantidade[i]))
            else:
                print(f"{item2} - esse medicamento não está cadastrado")   
        elif acao2 == "2":
            print("Dias disponíveis:")
            texto1 = set(dia)
            textoDias = "\n".join(texto1)
            print(textoDias)
            dia2 = input("Informe o dia para consultar: ")
            if dia2 in textoDias:
                print(f"{dia2}:")
                for i, dias1 in enumerate(dia):
                    if dia2 == dias1:
                        print("\t{} às {} - {} comprimidos".format(medicamento[i], hora[i], quantidade[i]))
            else:
                print(f"{dia2} não tem medicamento cadastrado")
    elif acao == "3":
        print("Medicamentos cadastrados:")
        for i in range(len(medicamento)):
            print("\t{}: {} às {} - {} comprimidos".format(medicamento[i], dia[i], hora[i], quantidade[i]))
    else:
        print("Comando inválido!\nTente novamente(1 - Cadastro| 2 - Consulta específica| 3 - Medicamentos cadastrados| 4-Sair:)")
    acao = input("1 - Cadastro| 2 - Consulta Específica| 3 - Medicamentos| 4-Sair:")
```
